## Function to call the Open AI API

```python
# Define prompts and variables
bmw_prompt = "I need to draft a section for the current ESRS report for training on business conduct of BMW. These are the measures they take: Training opportunities are continuously refined for specific target groups, For example, the company wide mandatory Compliance Essential training was expanded to six languages, Compliance training on with online and classroom training every two years, more than 81000 managers and staff have already completed training, 38000 have received anti trust compliance training, Mandatory data privacy training, which is available in 13 languages, Training on human rights. First, create a report section as a flow text. Second, provide some recommendations on what can be done to improve the compliance trainings"
db_prompt = "I need to draft a section for the current ESRS report for training on business conduct of Deutsche Bank. These are the measures they take: - training modules on communicating with clients, identifying and managing conflicts, and checking products’ suitability and appropriateness - module on Code of Conduct includes topics related to product responsibility as well - Deutsche Bank believes it is vital for all employees to complete this training - not participating in training can affect employees’ compensation and their manager’s - Employees in Investment Bank, Corporate Bank, and International Private Bank are required to complete online training module named “The Essentials of our Duties to Customers” or, depending on the Employee’s client base, a similar module that addresses consumer protection and module “The Essentials of Managing our Conflicts of Interest” - modules are done when joining the bank and then every two years - “The Essentials of Managing our Conflicts of Interest”: previously focused on business conflicts, was updated in recent years to include content on personal conflicts that was formerly part of a separate module; a new section on personal conflicts tailored to the needs of Infrastructure staff was added as well. First, create a report section as a flow text. Second, provide some recommendations on what can be done to improve the compliance trainings"
hannre_prompt = "I need to draft a section for the current ESRS report for training on business conduct of Hannover Re. These are the measures they take: - On joining the Group, all new members of staff take compliance training - In 2022, in-person training sessions were held for altogether 122 new employees at Hannover Head Office alone, new compliance training cycle was also launched for all staff in conjunction with implementation of the new Code of Conduct - all employees undergo compliance training in the Code of Conduct in a rotating cycle of three years - focused training activities are held on various compliance topics throughout the year for specific target groups - e.g. intranet portal and online newsletters to bring compliance-related issues to the attention of the workforce - Compliance unit and CCO keep employees informed about changes in legislation, if they affect their work. First, create a report section as a flow text. Second, provide some recommendations on what can be done to improve the compliance trainings"


prompts = [bmw_prompt, db_prompt, hannre_prompt]
models = ["gpt-3.5-turbo", "gpt-4"]
temperatures = [0.2, 0.8, 1.2]
examples = [True, False]


# Placeholder for storing results
results = []

# Function to call the API
def call_openai_api(model, temperature, prompt, example):
    """
    Calls the OpenAI API to generate text based on the provided prompt, model, and temperature settings.

    Args:
    model (str): The model to be used for the API call.
    temperature (float): The temperature setting for the API call.
    prompt (str): The prompt to be used for text generation.
    example (bool): Flag to determine if example-based learning is used.

    Returns:
    str: The generated text from the API.
    """
    if example == True: 
        try:
            response = client.chat.completions.create(
                model=model,
                messages=[
                    {"role": "system", "content": "You are an AI assistant experienced in ESRS guidelines and corporate sustainability reporting for large German companies."},
                    {"role": "user", "content": "I need to draft a section for the current ESRS report for training on business conduct of Airbus. Create a report section as a flow text."},
                    {"role": "assistant", "content": first_example},
                    {"role": "user", "content": "I need to draft a section for the current ESRS report for training on business conduct of Allianz. Create a report section as a flow text."},
                    {"role": "assistant", "content": second_example},
                    {"role": "user", "content": "I need to draft a section for the current ESRS report for training on business conduct of SAP. Create a report section as a flow text."},
                    {"role": "assistant", "content": third_example},
                    {"role": "user", "content": "I need to draft a section for the current ESRS report for training on business conduct of Siemens. Create a report section as a flow text."},
                    {"role": "assistant", "content": fourth_example},
                    {"role": "user", "content": "I need to draft a section for the current ESRS report for training on business conduct of Vonovia. Create a report section as a flow text."},
                    {"role": "assistant", "content": fifth_example},
                    {"role": "user", "content": prompt},
                ],
                temperature=temperature,
            )
            content = response.choices[0].message.content
        except Exception as e:
            content = str(e)
            
    else:
        try:
            response = client.chat.completions.create(
                model=model,
                messages=[
                    {"role": "system", "content": "You are an AI assistant experienced in ESRS guidelines and corporate sustainability reporting for large German companies."},
                    {"role": "user", "content": prompt},
                ],
                temperature=temperature,
            )
            content = response.choices[0].message.content
        except Exception as e:
            content = str(e) 
    
    return content

# Loop over all the combinations
for model in models:
    for temperature in temperatures:
        for example, technique in zip(examples, ["Few-Shot-Learning", "Zero-Shot-Learning"]):
            for prompt, company_name in zip(prompts, ["BMW", "Deutsche Bank", "Hannover RE"]):
                content = call_openai_api(model, temperature, prompt, example)
                # Store the results in the list
                results.append({
                    "Company": company_name,
                    "model": model,
                    "technique": technique,
                    "temperature": temperature,
                    "prompt": prompt,
                    "content": content
                })

# Convert the results to a dataframe
df_results = pd.DataFrame(results)

# Store dataframe in excel
df_results.to_excel("results.xlsx")
```
